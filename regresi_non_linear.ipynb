```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path='china_gdp.csv'
df = pd.read_csv(path)
df.head()
```

Persamaan Linear

```python
x = np.arange(-5.0, 5.0, 0.4) # untuk membuat sebuah array yang berisi urutan bilangan dari -5.0 hingga 4.8, dengan selisih antar bilangan sebesar 0.4.

# y = ax + b 
# Contoh y = 2x + 3
y = 2*(x) + 3

y_random = 2 * np.random.normal(size=x.size) #  menghasilkan array nilai acak sebesar array x dengan distribusi normal 
ydata = y + y_random
plt.plot(x, ydata, 'bo') #  'bo' menentukan bagaimana plot akan ditampilkan, yaitu dengan titik biru ('b' singkatan dari blue dan 'o' singkatan dari circle).

plt.plot(x,y, 'r') # membuat grafik dengan warna 'r' red
plt.ylabel('Y')
plt.xlabel('X')


plt.show()
```

Persamaan Non-Linier (polynomial)

y = ax3 + bx2 + cx + d

Contoh: y = x^3 + 2x^2 + 3x + 4

```python
x = np.arange(-5.0, 5.0, 0.4)
y = 1 *(x**3) + 2* (x**2) + 3*x + 4

y_random = 20 * np.random.normal(size=x.size) #  menghasilkan array nilai acak sebesar array x dengan distribusi normal 
ydata = y + y_random
plt.plot(x, ydata, 'bo') #  'bo' menentukan bagaimana plot akan ditampilkan, yaitu dengan titik biru ('b' singkatan dari blue dan 'o' singkatan dari circle).

plt.plot(x,y, 'r') # membuat grafik dengan warna 'r' red
plt.ylabel('Y')
plt.xlabel('X')


plt.show()
```

Persamaan Non-Linier (Quadratic)
y = x^2

```python
x = np.arange(-5.0, 5.0, 0.4)
y = np.power(x,2) # x^2

y_random = 2 * np.random.normal(size=x.size) #  menghasilkan array nilai acak sebesar array x dengan distribusi normal 
ydata = y + y_random
plt.plot(x, ydata, 'bo') #  'bo' menentukan bagaimana plot akan ditampilkan, yaitu dengan titik biru ('b' singkatan dari blue dan 'o' singkatan dari circle).

plt.plot(x,y, 'r') # membuat grafik dengan warna 'r' red
plt.ylabel('Y')
plt.xlabel('X')


plt.show()
```

Persamaan Non-Linier (Exponential)
y = e^x

```python
x = np.arange(-5.0, 5.0, 0.4)
y = np.exp(x) # 

y_random = 10 * np.random.normal(size=x.size) #  menghasilkan array nilai acak sebesar array x dengan distribusi normal 
ydata = y + y_random
plt.plot(x, ydata, 'bo') #  'bo' menentukan bagaimana plot akan ditampilkan, yaitu dengan titik biru ('b' singkatan dari blue dan 'o' singkatan dari circle).

plt.plot(x,y, 'r') # membuat grafik dengan warna 'r' red
plt.ylabel('Y')
plt.xlabel('X')


plt.show()
```

Persamaan Non-Linier (Logarithmic)
y = log x

```python
x = np.arange(0.1, 10.0, 0.4)
y = np.log(x) # 

y_random = 0.5 * np.random.normal(size=x.size) #  menghasilkan array nilai acak sebesar array x dengan distribusi normal 
ydata = y + y_random
plt.plot(x, ydata, 'bo') #  'bo' menentukan bagaimana plot akan ditampilkan, yaitu dengan titik biru ('b' singkatan dari blue dan 'o' singkatan dari circle).

plt.plot(x,y, 'r') # membuat grafik dengan warna 'r' red
plt.ylabel('Y')
plt.xlabel('X')


plt.show()
```

Persamaan Non-Linier (Sigmoidal/Logistic)
Y = a+(b/1+c^(x-d))

```python
# sigmoidal/Logistic
x = np.arange(-5.0, 8.0, 0.4)
y = 1+(1/(1+np.power(4,x-2)))

y_random = 0.1 * np.random.normal(size=x.size) #  menghasilkan array nilai acak sebesar array x dengan distribusi normal 
ydata = y + y_random
plt.plot(x, ydata, 'bo') #  'bo' menentukan bagaimana plot akan ditampilkan, yaitu dengan titik biru ('b' singkatan dari blue dan 'o' singkatan dari circle).

plt.plot(x,y, 'r') # membuat grafik dengan warna 'r' red
plt.ylabel('Y')
plt.xlabel('X')


plt.show()
```

Persamaan Non-Linier (Sigmoidal/Logistic) (+x)

```python
# sigmoidal/Logistic (+x)
x = np.arange(-5.0, 8.0, 0.4)
y = 1.0/(1.0 +np.exp(x))

y_random = 0.05 * np.random.normal(size=x.size) #  menghasilkan array nilai acak sebesar array x dengan distribusi normal 
ydata = y + y_random
plt.plot(x, ydata, 'bo') #  'bo' menentukan bagaimana plot akan ditampilkan, yaitu dengan titik biru ('b' singkatan dari blue dan 'o' singkatan dari circle).

plt.plot(x,y, 'r') # membuat grafik dengan warna 'r' red
plt.ylabel('Y')
plt.xlabel('X')


plt.show()
```
