```python
import pandas as pd # untuk melakukan processing dan analysis data 
import numpy as np # library python yg berfokus pada scientific computing
import matplotlib.pyplot as plt #untuk visualisasi data

base_dir = "automobileEDA.csv" # deklarasi file dataset kita
df = pd.read_csv(base_dir) # untuk membaca file dataset

df.head() # menampilkan 5 baris data pertama pada dataset kita
```

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression() # Deklarasi model Linear Regression
```

```python
# Rumus model linear y = a + bx

X_par = df[['engine-size']].values # set x variabel independen/ explanatory variabel
Y_par = df[['price']].values # set y variabel dependen/ respon variabel => yg mau kita prediksi

lm.fit(X_par,Y_par) # melatih model dengan data

a = lm.intercept_ # ambil data intercept
b = lm.coef_ # ambil data slop

print(a, b) # cetak data intercept dan slop 
```

```python
ukm = np.arange(500,1001) # data yang dihasilkan berbentuk 1 dimensi
ukm
```

```python
um = ukm.reshape(-1,1) # mengubah data menjadi 2 dimensi
um.shape
um
```

```python
# buat function untuk menghitung prediksi_harga dengan mneggunakan persamaan formula regression
def model_linier (x,a,b):
  ymodel = a + b * x
  return ymodel
```

```python
# cara 1
prediksi_harga1 = model_linier(um,a,b) # menghitung prediksi_harga berdasarkan function model_linier yang telah dibuat
prediksi_harga1 
```

```python
# cara 2
prediksi_harga2 = lm.predict(um) # menghitung prediksi_harga 
prediksi_harga2
```

```python
plt.figure(figsize = (8,8)) # mengatur ukuran visualisasi gambar


# visualisasi data dimana ukm berkorelasi dengan sumbu x dan prediksi_harga2 berkorelasi dengan sumbu y
plt.plot(ukm,prediksi_harga2, linewidth=2.0, label='Prediksi harga') 

plt.legend(loc='best') # atur posisi label
plt.xlabel('Engine Size') # penamaan label disumbu x
plt.ylabel('Price') # penamaan label disumbu y

plt.show() # menampilkan visualisasi data
```
